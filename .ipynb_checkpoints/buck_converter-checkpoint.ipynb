```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simulation parameters
Vin = 12.0          # Input voltage (V)
Vref = 5.0          # Target output voltage (V)
fs = 50000          # Switching frequency (Hz)
Ts = 1 / fs         # Time step (s)
Tsim = 0.02         # Simulation time (s)
steps = int(Tsim / Ts)

# Buck converter components
L = 220e-6          # Inductor (H)
C = 470e-6          # Capacitor (F)
R_normal = 5.0      # Normal load (ohm, ~1A load)
R_overload = 1.67   # Overload (ohm, ~3A load)

print("Parameters set. Vin=12V, Target=5V, Normal Load=5Ω, Overload=1.67Ω")
```

```python
def simulate_buck(Kp=0.0, Ki=0.0, R_load=5.0, Tsim_local=None, anti_windup=True):
    """
    Simulate averaged buck converter for given PI gains.
    Returns time, output voltage, duty, inductor current.
    """
    Tsim_use = Tsim if Tsim_local is None else Tsim_local
    steps_local = int(Tsim_use / Ts)
    iL = 0.0
    Vc = 0.0
    integral = 0.0
    time, Vout, Duty, IL = [], [], [], []

    for n in range(steps_local):
        error = Vref - Vc

        # Integrator with anti-windup
        if Ki != 0:
            tentative_D = Kp*error + Ki*integral
            tentative_D = max(0.0, min(1.0, tentative_D))
            if (not anti_windup) or (0.001 < tentative_D < 0.999):
                integral += error * Ts

        # Controller output
        D = Kp*error + Ki*integral
        D = max(0.0, min(1.0, D))

        # Buck dynamics
        Vswitch = D * Vin
        diL = (Vswitch - Vc) / L * Ts
        iL += diL
        dVc = (iL - Vc / R_load) / C * Ts
        Vc += dVc

        # Save
        time.append(n*Ts)
        Vout.append(Vc)
        Duty.append(D)
        IL.append(iL)

    return np.array(time), np.array(Vout), np.array(Duty), np.array(IL)
```

```python
def find_Ku_Pu(Kp_start=0.1, Kp_end=300.0, Kp_steps=80, sim_time=0.05, osc_threshold=0.02):
    """
    Sweep Kp with P-only control to find Ku and Pu.
    """
    Kp_values = np.linspace(Kp_start, Kp_end, Kp_steps)
    Ku, Pu = None, None

    for Kp_try in Kp_values:
        t, vout, _, _ = simulate_buck(Kp=Kp_try, Ki=0.0, R_load=R_normal, Tsim_local=sim_time, anti_windup=False)
        segment = vout[len(vout)//2:]
        mean_val = np.mean(segment)
        centered = segment - mean_val

        # detect zero-crossings
        sign_changes = np.where(np.diff(np.sign(centered)) != 0)[0]

        if len(sign_changes) >= 6:
            zero_spacings = np.diff(sign_changes) * Ts
            if len(zero_spacings) > 0:
                half_period = np.mean(zero_spacings)
                period = 2 * half_period
                p2p = np.max(segment) - np.min(segment)
                if p2p > osc_threshold:
                    Ku, Pu = Kp_try, period
                    break
    return Ku, Pu

Ku, Pu = find_Ku_Pu()
print(f"Ku = {Ku}, Pu = {Pu}")
```

```python
if Ku is None:
    print("Ultimate gain Ku not found. Try increasing Kp_end or sim_time in Cell 4.")
else:
    Kp_ZN = 0.45 * Ku
    Ki_ZN = Kp_ZN * 1.2 / Pu
    print(f"Ziegler–Nichols PI gains: Kp = {Kp_ZN:.4f}, Ki = {Ki_ZN:.4f}")
```

```python
if Ku is None:
    raise RuntimeError("PI tuning failed. Adjust search parameters in Cell 4.")

t, vout, duty, il = simulate_buck(Kp=Kp_ZN, Ki=Ki_ZN, R_load=R_normal, Tsim_local=Tsim)

plt.figure(figsize=(8,4))
plt.plot(t, vout, label="Vout (regulated)")
plt.axhline(Vref, color='r', linestyle='--', label="Target 5V")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Closed-Loop with ZN PI (Normal Load)")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(t, duty, label="Duty cycle")
plt.xlabel("Time (s)")
plt.ylabel("Duty")
plt.title("Duty Adjustment by PI")
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(t, il, label="Inductor Current")
plt.axhline(3.0, color='r', linestyle='--', label="Overcurrent Limit")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Inductor Current (Normal Load)")
plt.legend()
plt.grid()
plt.show()
```

```python
limit_I = 3.0   # A
limit_V = 5.5   # V

t, vout, duty, il = simulate_buck(Kp=Kp_ZN, Ki=Ki_ZN, R_load=R_overload, Tsim_local=Tsim)

# Fault detection
fault = "Normal"
if np.max(il) > limit_I:
    fault = "Overcurrent"
elif np.max(vout) > limit_V:
    fault = "Overvoltage"

if fault == "Normal":
    fault = "None"
plt.figure(figsize=(8,4))
plt.plot(t, vout, label="Vout (overload)")
plt.axhline(limit_V, color='r', linestyle='--', label="Overvoltage limit")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Overload Response with PI Control")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(t, il, label="Inductor Current (A)")
plt.axhline(limit_I, color='r', linestyle='--', label="Overcurrent Limit")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Inductor Current in Overload Case")
plt.legend()
plt.grid()
plt.show()

print("Fault detected:", fault)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation parameters
Vin = 12.0          # Input voltage (V)
Vref = 5.0          # Target output voltage (V)
fs = 50000          # Switching frequency (Hz)
Ts = 1 / fs         # Time step (s)
Tsim = 0.02         # Simulation time (s)
steps = int(Tsim / Ts)

# Buck converter components
L = 220e-6          # Inductor (H)
C = 470e-6          # Capacitor (F)
R_normal = 5.0      # Normal load (ohm, ~1A load)
R_overload = 1.67   # Overload (ohm, ~3A load)

print("Parameters set. Vin=12V, Target=5V, Normal Load=5Ω, Overload=1.67Ω")

def simulate_buck(Kp, Ki, R_load, Tsim_local=None, anti_windup=True):
    """
    Simulate averaged buck converter for given PI gains.
    Returns time, output voltage, duty, inductor current.
    """
    Tsim_use = Tsim if Tsim_local is None else Tsim_local
    steps_local = int(Tsim_use / Ts)
    iL = 0.0
    Vc = 0.0
    integral = 0.0
    time, Vout, Duty, IL = [], [], [], []

    for n in range(steps_local):
        error = Vref - Vc

        # Controller output
        D = Kp*error + Ki*integral
        
        # Anti-windup and clamping
        if anti_windup:
            tentative_D = D
            D = max(0.0, min(1.0, D))
            if D < 1.0 and D > 0.0:
                 integral += error * Ts
            elif D == 1.0 and error > 0:
                 integral += error * Ts
            elif D == 0.0 and error < 0:
                 integral += error * Ts
        else:
            integral += error * Ts
            D = max(0.0, min(1.0, D))

        # Buck dynamics
        Vswitch = D * Vin
        diL = (Vswitch - Vc) / L * Ts
        iL += diL
        dVc = (iL - Vc / R_load) / C * Ts
        Vc += dVc

        # Save
        time.append(n*Ts)
        Vout.append(Vc)
        Duty.append(D)
        IL.append(iL)

    return np.array(time), np.array(Vout), np.array(Duty), np.array(IL)

# --- Manual PI Gain Tuning (Replaces ZN) ---
# Kp: Controls the speed of response. Higher Kp = faster, more oscillatory.
# Ki: Eliminates steady-state error. Higher Ki = faster error elimination.
# Tuned for a well-damped response.
Kp_tuned = 10.0
Ki_tuned = 500.0
print(f"Manually tuned PI gains: Kp = {Kp_tuned:.4f}, Ki = {Ki_tuned:.4f}")

# --- Normal Load Simulation ---
t_norm, vout_norm, duty_norm, il_norm = simulate_buck(Kp=Kp_tuned, Ki=Ki_tuned, R_load=R_normal)

plt.figure(figsize=(8,4))
plt.plot(t_norm, vout_norm, label="Vout (regulated)")
plt.axhline(Vref, color='r', linestyle='--', label="Target 5V")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Closed-Loop with Manually Tuned PI (Normal Load)")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(t_norm, duty_norm, label="Duty cycle")
plt.xlabel("Time (s)")
plt.ylabel("Duty")
plt.title("Duty Adjustment by PI")
plt.grid()
plt.legend()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(t_norm, il_norm, label="Inductor Current")
plt.axhline(3.0, color='r', linestyle='--', label="Overcurrent Limit")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Inductor Current (Normal Load)")
plt.legend()
plt.grid()
plt.show()

# --- Overload Simulation and Fault Detection ---
limit_I = 3.0   # A
limit_V = 5.5   # V

t_ovl, vout_ovl, duty_ovl, il_ovl = simulate_buck(Kp=Kp_tuned, Ki=Ki_tuned, R_load=R_overload)

# Fault detection logic
fault = "None"
if np.max(il_ovl) > limit_I:
    fault = "Overcurrent"
elif np.max(vout_ovl) > limit_V:
    fault = "Overvoltage"

plt.figure(figsize=(8,4))
plt.plot(t_ovl, vout_ovl, label="Vout (overload)")
plt.axhline(limit_V, color='r', linestyle='--', label="Overvoltage limit")
plt.xlabel("Time (s)")
plt.ylabel("Voltage (V)")
plt.title("Overload Response with PI Control")
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(8,4))
plt.plot(t_ovl, il_ovl, label="Inductor Current (A)")
plt.axhline(limit_I, color='r', linestyle='--', label="Overcurrent Limit")
plt.xlabel("Time (s)")
plt.ylabel("Current (A)")
plt.title("Inductor Current in Overload Case")
plt.legend()
plt.grid()
plt.show()

print("Fault detected:", fault)
```
